# Arrays in Python

- Lists
- Tuples
- Sets

#### Lists :
- Ordered
- Allows duplicates
- Store different data types

```python
list = [1,"two",3,4.0]

# First element have zero index
```

```python
print(list[0])
```

#### Sets :
- Unordered
- Unique elements
- Store different data types

```python
set = {1,2.0,"three",4}

# Cannot add another 1 - 2.0 - "three" string or 4 elements in same set
```

#### Tuples :
- Ordered
- Elements cannot be modified
- Allows duplicates
- Store different data types

It's like upgraded lists

```python
tuple = (1,2,3.333333,True)
```
